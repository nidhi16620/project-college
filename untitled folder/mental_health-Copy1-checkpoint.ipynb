```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv("SMH.csv")
df
```

```python
from sklearn.preprocessing import LabelEncoder
lt=LabelEncoder()
```

```python
df['Choose your gender']=lt.fit_transform(df['Choose your gender'])
df['Do you have Depression?']=lt.fit_transform(df['Do you have Depression?'])
df['Do you have Anxiety?']=lt.fit_transform(df['Do you have Anxiety?'])
df['Do you have Panic attack?']=lt.fit_transform(df['Do you have Panic attack?'])
df['Did you seek any specialist for a treatment?']=lt.fit_transform(df['Did you seek any specialist for a treatment?'])
df
```

```python
df.isnull().sum()
```

```python
df.fillna({'Age':'0'},inplace=True)
df.isnull().sum()
```

```python
df.corr()
```

```python

x= df.iloc[:,[2,5,8,9,10]].values
y= df.iloc[:,[7]].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
lo= LogisticRegression()
lo.fit(x_train,y_train)
```

```python
pred= lo.predict(x_test)
pred
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
re= confusion_matrix(y_test,pred)
print(re)
print("accurcy",accuracy_score(y_test,pred))
```
